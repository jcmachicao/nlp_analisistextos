```python
import PyPDF2
import pandas as pd
import re
```

```python
ruta = 'D://trabajos/'
arch = 'archivoejemplo.pdf'
```

```python
tex_acum = ''

fileObj = open(ruta + arch, 'rb')     
pdfRead = PyPDF2.PdfFileReader(fileObj)
pag = pdfRead.numPages

for i in range(0, pag):
    page = pdfRead.getPage(i)
    lector = page.extractText()
    texto = str(lector)
    tex_acum += ' ' + texto
print(tex_acum)
```
